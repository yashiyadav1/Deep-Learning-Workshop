```python
import numpy as np
import matplotlib.pyplot as plt

# Simulating a dataset where y values are exam scores between 0 and 100
np.random.seed(42)  # Ensuring reproducibility

# Assuming students study between 0 and 20 hours for the exam
X = np.random.rand(100, 1) * 20  # Study hours

# Generating exam scores such that they somewhat depend on study hours
# Assuming the base score is 50 (without studying), and every hour of studying adds 2.5 points on average
# Also adding some random noise
y = 50 + 2.5 * X + np.random.randn(100, 1) * 10  # Exam scores

# Ensuring that all scores are within 0 to 100
y = np.clip(y, 0, 100)

# Plotting the dataset
plt.scatter(X, y)
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.title("Study Hours vs Exam Score")
plt.show()

# Building the Linear Regression model
X_b = np.c_[np.ones((100, 1)), X]  # add x0 = 1 to each instance
learning_rate = 0.0001  # smaller learning rate
n_iterations = 1000
m = len(X_b)

theta = np.random.randn(2, 1)  # random initialization

for iteration in range(n_iterations):
    gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
    theta -= learning_rate * gradients

# Making predictions
X_new = np.array([[0], [20]])  # From 0 to 20 study hours
X_new_b = np.c_[np.ones((2, 1)), X_new]  # add x0 = 1 to each instance
y_predict = X_new_b.dot(theta)
y_predict = np.clip(y_predict, 0, 100)  # Clip predictions

# Plotting predictions
plt.plot(X_new, y_predict, "r-", linewidth=2, label="Predictions")
plt.plot(X, y, "b.")
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.title("Predicted vs Actual Scores")
plt.legend()
plt.show()
```
